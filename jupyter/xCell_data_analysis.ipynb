```python
fig = plt.figure(figsize=(14,7))
ax1 = fig.add_subplot(111)
ax1.bar(x = np.arange(df.shape[0]), height=np.mean(df, axis=1))
ax1.set_xticks(np.arange(df.shape[0]))
ax1.set_xticklabels(df.index, rotation=90)
plt.show()
```

```python
for ctype in df.index:
    fig = plt.figure(figsize=(8,6))
    ax1 = fig.add_subplot(111)
    ax1.hist(df.loc[ctype], bins=30)
    ax1.set_title(ctype)
    plt.show()
```
